```python
#MOD26 & 13
import string

def rot13(text):
    rot13_table = str.maketrans(
        string.ascii_lowercase + string.ascii_uppercase,
        string.ascii_lowercase[13:] + string.ascii_lowercase[:13] +
        string.ascii_uppercase[13:] + string.ascii_uppercase[:13]
    )
    return text.translate(rot13_table)


encrypted_text = "cvpbPGS{abg_gbb_onq_bs_n_ceboyrz}"
print("Encrypted text:", encrypted_text)
decrypted_text = rot13(encrypted_text)
print("Decrypted text:", decrypted_text)
```

```python
#new caesar
import string

LOWERCASE_OFFSET = ord("a")
ALPHABET = string.ascii_lowercase[:16]

def b16_decode(solve):
    dec = ""
    for index in range(0, len(solve), 2):
        first = solve[index]
        second = solve[index + 1]
        first = ALPHABET.index(first)
        second = ALPHABET.index(second)
        firstB = "{0:04b}".format(first)
        secondB = "{0:04b}".format(second)
        decB = int(firstB + secondB, 2)
        dec += chr(decB)
    return dec

def shift(c, k):
	t1 = ord(c) + LOWERCASE_OFFSET
	t2 = ord(k) + LOWERCASE_OFFSET
	return ALPHABET[(t1 - t2) % len(ALPHABET)]

def contains_only_ascii(arr):
    for char in arr:
        if not (32 <= ord(char) <= 126):
            return False
    return True

flag = "kjlijdliljhdjdhfkfkhhjkkhhkihlhnhghekfhmhjhkhfhekfkkkjkghghjhlhghmhhhfkikfkfhm"
for key in ALPHABET:
	enc = ""
	for i, c in enumerate(flag):
		enc += shift(c, key)
	d=b16_decode(enc)
	if contains_only_ascii(d):
		print(d)
		
```

```python
#caesar
def decrypt_caesar_cipher(text, shift):
    decrypted_text = ""
    
    for char in text:
        if char.isalpha():
            shift_amount = shift % 26
            if char.islower():
                decrypted_text += chr((ord(char) - 97 - shift_amount) % 26 + 97)
            elif char.isupper():
                decrypted_text += chr((ord(char) - 65 - shift_amount) % 26 + 65)
        else:
            decrypted_text += char

    return decrypted_text

def break_caesar_cipher(text):
    for shift in range(1, 26):
        possible_plaintext = decrypt_caesar_cipher(text, shift)
        print(f"Shift {shift}: {possible_plaintext}")

ciphered_text = "dspttjohuifsvcjdpoabrkttds"
break_caesar_cipher(ciphered_text)
```
